```python
#Q6 Ans:
def modify_objects(mutable_arg, immutable_arg):
    print("Before modification:")
    print("Mutable argument:", mutable_arg)
    print("Immutable argument:", immutable_arg)

    # Modifying the mutable argument (e.g., a list)
    mutable_arg.append(100)

    # Modifying the immutable argument (e.g., an integer)
    immutable_arg += 10

    print("\nAfter modification:")
    print("Mutable argument:", mutable_arg)
    print("Immutable argument:", immutable_arg)


# Example of mutable and immutable arguments
my_list = [1, 2, 3]
my_int = 50

# Call the function
modify_objects(my_list, my_int)

# Print the original arguments outside the function
print("\nOutside the function:")
print("Original list:", my_list)
print("Original integer:", my_int)
```

```python
#Q8 Ans:
def logical_operators_demo(a, b):
    # Using `and` operator
    print(f"a = {a}, b = {b}")
    print(f"Using 'and': {a > 0 and b > 0}")  # True if both a and b are greater than 0

    # Using `or` operator
    print(f"Using 'or': {a > 0 or b > 0}")  # True if at least one of a or b is greater than 0

    # Using `not` operator
    print(f"Using 'not' on a > 0: {not (a > 0)}")  # True if a is not greater than 0
    print(f"Using 'not' on b > 0: {not (b > 0)}")  # True if b is not greater than 0


# Example with values
logical_operators_demo(5, -3)
logical_operators_demo(-5, 3)
logical_operators_demo(0, 0)
```

```python
#Q9 Ans:
def convert_input():
    # Take user input as a string
    user_input = input("Enter a value: ")

    # Convert to integer
    try:
        int_value = int(user_input)
        print(f"Converted to Integer: {int_value}")
    except ValueError:
        print("Could not convert to Integer.")

    # Convert to float
    try:
        float_value = float(user_input)
        print(f"Converted to Float: {float_value}")
    except ValueError:
        print("Could not convert to Float.")

    # Convert to boolean
    # For boolean, if input is 'True', '1', or any non-empty string, it's True.
    # Otherwise, it's False.
    if user_input.lower() == 'true' or user_input == '1' or user_input:
        bool_value = True
    else:
        bool_value = False
    print(f"Converted to Boolean: {bool_value}")

# Call the function
convert_input()
```

```python
#Q14 Ans
string = input("Enter a string: ")
reversed_string = ""

index = len(string) - 1

# Using while loop to reverse the string
while index >= 0:
    reversed_string += string[index]
    index -= 1
print("Reversed string:", reversed_string)
```

```python
#Q15 Ans
num = int(input("Enter a number to calculate its factorial: "))
factorial = 1
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    # Using while loop to calculate the factorial
    i = 1
    while i <= num:
        factorial *= i
        i += 1
    print(f"The factorial of {num} is {factorial}.")
```
